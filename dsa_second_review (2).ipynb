## eCommerce Website Inventory Management

### Introduction
Efficient inventory management is crucial for eCommerce businesses to ensure product availability while minimizing costs. This system provides a structured approach to managing stock, tracking sales, and optimizing inventory replenishment using a **Binary Search Tree (BST)** and **Object-Oriented Programming (OOP) principles**.

### Objectives
- Implement a **scalable** inventory management system for an eCommerce platform.
- Ensure **efficient stock management** with real-time updates.
- Provide **fast retrieval** of product details using **BST for search operations**.
- Optimize stock replenishment and order fulfillment processes.

### System Architecture
The system consists of the following key components:
1. **Product Class:** Represents individual products with attributes like ID, name, category, price, and stock.
2. **BST-based Inventory:** Stores product data in a Binary Search Tree for efficient searching, insertion, and deletion.
3. **Cart System:** Allows customers to add/remove items and proceed to checkout.
4. **Sales Tracking Module:** Simulates purchase transactions and updates stock levels.
5. **Visualization Module:** Uses **matplotlib** to provide insights into inventory trends.

### Data Structure Used & Justification
**Binary Search Tree (BST) for Inventory Management**
- **Efficient Searching:** BST allows **O(log n) search time**, ensuring quick retrieval of products.
- **Ordered Storage:** An **in-order traversal** provides a sorted list of products based on ID.
- **Balanced Performance:** Optimized to maintain efficient insertions and deletions over time.

### Step-by-Step Flow of the Project
1. **Initialize Inventory** – Create products and store them in a BST.
2. **Search Products** – Retrieve product details quickly.
3. **Manage Stock** – Add new products, update stock levels, or delete discontinued items.
4. **Cart Operations** – Customers add/remove products before checkout.
5. **Process Sales** – Deduct purchased items from stock and track transactions.
6. **Restock Alerts** – Identify low-stock products and trigger alerts for replenishment.
7. **Data Visualization** – Display inventory trends using bar charts.

### **Explanation of Code Structure & Design**  

The system follows a modular and object-oriented approach, ensuring maintainability, scalability, and efficiency. It consists of well-defined **classes and modules**, each playing a specific role in managing the inventory system using a **Binary Search Tree (BST)**.

### **Key Components of the System**  

### 1. **Product Class**  
- Represents a single product in the inventory.  
- Encapsulates essential attributes such as:  
  - `product_id`: A unique identifier for each product.  
  - `name`: The name of the product.  
  - `price`: The cost per unit of the product.  
  - `stock`: The available quantity of the product.  
- Methods in this class include getters and setters for accessing and updating product details.

### 2. **Node Class (BST Node Representation)**  
- Represents a **single node** in the **Binary Search Tree (BST)**.  
- Contains:  
  - `product`: An instance of the `Product` class.  
  - `left`: A pointer to the left child (for products with lower IDs).  
  - `right`: A pointer to the right child (for products with higher IDs).  

### 3. **InventoryBST Class (Binary Search Tree for Inventory Management)**  
- Implements a **BST structure** to efficiently store and retrieve products.  
- Supports **core operations** such as:  
  - **Insertion**: Adds new products to the BST based on their `product_id`.  
  - **Search**: Finds a product by `product_id` in O(log n) time complexity.  
  - **Deletion**: Removes a product while maintaining BST structure.  
  - **Traversal**: Performs in-order traversal to list products in sorted order.

### 4. **InventorySystem Class (Manages Inventory Operations)**  
- Acts as a higher-level interface between the BST-based inventory and the user.  
- Facilitates user-friendly functions for:  
  - Adding/removing/searching products.  
  - Displaying available products.  
  - Managing stock levels.

### 5. **Cart Class (Handles Shopping Cart Operations)**  
- Allows users to add/remove products from their cart.  
- Maintains a dictionary with `product_id` as keys and quantity as values.  
- Computes total price at checkout, applying discounts if needed.

### **Adherence to Object-Oriented Programming (OOP) Concepts**  

The system follows fundamental **OOP principles**, ensuring modularity, reusability, and scalability.  

### **1. Encapsulation**  
- **Definition**: Encapsulation refers to **hiding data** and exposing only necessary functions.  
- **Implementation**:  
  - Each class encapsulates its own data and operations.  
  - The `Product` class keeps attributes private and provides getter and setter methods.  
  - The BST implementation restricts direct access to node manipulation, exposing only search, insert, and delete operations.

### **2. Polymorphism**  
- **Definition**: Polymorphism allows **different implementations of the same method** across multiple classes.  
- **Implementation**:  
  - If there are multiple inventory categories, methods like `search()` or `display()` can be overridden for different implementations.

### **3. Abstraction**  
- **Definition**: Abstraction **hides complexity** and provides only essential functionalities.  
- **Implementation**:  
  - The `InventorySystem` class hides **BST complexities**, exposing user-friendly functions for product management.  
  - Users interact with simple methods (`add_product()`, `find_product()`), without dealing with low-level BST operations.


```python
# -------------------------------
# Domain Classes and Data Structures
# -------------------------------

class Product:
    def __init__(self, id, name, category, quantity, price):
        # Initialize a Product object with id, name, category, quantity, and price.
        self.id = id                  # Unique identifier for the product.
        self.name = name              # Name of the product.
        self.category = category      # Category of the product.
        self.quantity = quantity      # Available quantity in stock.
        self.price = price            # Price per unit of the product.

    def __str__(self):
        # Return a formatted string representation of the product.
        return (f"ID: {self.id}, Name: {self.name}, Category: {self.category}, "
                f"Quantity: {self.quantity}, Price: ${self.price:.2f}")


class Node:
    def __init__(self, product): 
        # Initialize a BST node that will hold a product.
        self.product = product        # The product stored in this node.
        self.left = None              # Pointer to the left child (products with lower IDs).
        self.right = None             # Pointer to the right child (products with higher IDs).


class InventoryBST:
    def __init__(self):
        # Initialize an empty Binary Search Tree for managing products.
        self.root = None              # Root node of the BST; initially, there is no product.

    def insert(self, product):
        """Insert a product into the BST."""
        # Call the recursive insertion helper, starting from the root.
        self.root = self._insert_rec(self.root, product)

    def _insert_rec(self, root, product):
        # Recursive helper function to insert a product into the BST.
        if root is None:
            # If there is no node here, create a new node for the product.
            return Node(product)
        if product.id < root.product.id:
            # If the product's id is less than the current node's id, insert into the left subtree.
            root.left = self._insert_rec(root.left, product)
        elif product.id > root.product.id:
            # If the product's id is greater than the current node's id, insert into the right subtree.
            root.right = self._insert_rec(root.right, product)
        # Return the (possibly unchanged) current node.
        return root

    def inorder(self):
        """Return products in sorted order (by product id)."""
        products = []                # Initialize an empty list to store products.
        self._inorder_rec(self.root, products)  # Fill the list using in-order traversal.
        return products              # Return the sorted list.

    def _inorder_rec(self, root, products):
        # Recursive helper function to perform an in-order traversal of the BST.
        if root:
            self._inorder_rec(root.left, products)  # Traverse the left subtree.
            products.append(root.product)           # Visit the current node.
            self._inorder_rec(root.right, products) # Traverse the right subtree.

    def search_by_id(self, id):
        """Search for a product by id."""
        # Start the recursive search from the root.
        return self._search(self.root, id)

    def _search(self, root, id):
        # Recursive helper function to search for a product by id.
        if root is None or root.product.id == id:
            # If the node is None or the current product matches, return the node.
            return root
        if id < root.product.id:
            # If the search id is less, go to the left subtree.
            return self._search(root.left, id)
        # Otherwise, search in the right subtree.
        return self._search(root.right, id)

    def search_by_category(self, category):
        """Search for products matching a category (case-insensitive)."""
        results = []                # List to store matching products.
        # Start recursive search after converting category to lowercase.
        self._search_by_category_rec(self.root, category.lower(), results)
        return results              # Return all products found in the category.

    def _search_by_category_rec(self, root, category, results):
        # Recursive helper function to search for products by category.
        if root:
            self._search_by_category_rec(root.left, category, results)  # Check left subtree.
            if root.product.category.lower() == category:
                # If the product's category matches (case-insensitive), add it to results.
                results.append(root.product)
            self._search_by_category_rec(root.right, category, results) # Check right subtree.

    def search_by_name(self, name):
        """Search for products containing a name substring (case-insensitive)."""
        results = []                # List to store matching products.
        # Begin recursive search after converting name to lowercase.
        self._search_by_name_rec(self.root, name.lower(), results)
        return results              # Return all products matching the name substring.

    def _search_by_name_rec(self, root, name, results):
        # Recursive helper function to search for products by name substring.
        if root:
            self._search_by_name_rec(root.left, name, results)  # Search left subtree.
            if name in root.product.name.lower():
                # If the substring is found in the product's name, add to results.
                results.append(root.product)
            self._search_by_name_rec(root.right, name, results) # Search right subtree.

    def delete(self, id):
        """Delete a product by id from the BST."""
        # Call the recursive deletion helper starting from the root.
        self.root = self._delete_rec(self.root, id)

    def _delete_rec(self, root, id):
        # Recursive helper function to delete a product by id.
        if root is None:
            # If the node is None, the product is not found.
            return root
        if id < root.product.id:
            # If the id to delete is less than the current node's id, go left.
            root.left = self._delete_rec(root.left, id)
        elif id > root.product.id:
            # If the id to delete is greater than the current node's id, go right.
            root.right = self._delete_rec(root.right, id)
        else:
            # Node with the matching id found.
            if root.left is None:
                # If there is no left child, return the right child (or None).
                return root.right
            elif root.right is None:
                # If there is no right child, return the left child.
                return root.left
            # If the node has two children, find the inorder successor.
            min_value = self._min_value(root.right)
            # Replace the product in this node with the inorder successor.
            root.product = min_value
            # Delete the inorder successor from the right subtree.
            root.right = self._delete_rec(root.right, min_value.id)
        return root

    def _min_value(self, root):
        # Helper function to find the minimum value node (leftmost node) in a subtree.
        while root.left:
            root = root.left  # Move to the left child.
        return root.product  # Return the product stored in the leftmost node.

# -------------------------------
# Inventory System with Cart and Menu
# -------------------------------

class InventorySystem:
    def __init__(self):
        # Initialize the inventory system with a BST for inventory and an empty cart.
        self.inventory = InventoryBST()  # Create the BST instance.
        self.cart = []                   # Initialize the shopping cart as an empty list.
        self.initialize_inventory()      # Populate the inventory with default products.

    def initialize_inventory(self):
        """Initialize inventory with default products."""
        # List of default products to add to the inventory.
        products = [
            Product(1, "Laptop", "Electronics", 10, 999.99),
            Product(2, "Smartphone", "Electronics", 20, 699.99),
            Product(3, "Headphones", "Electronics", 30, 149.99),
            Product(4, "T-shirt", "Clothing", 50, 19.99),
            Product(5, "Jeans", "Clothing" , 25, 39.99),
            Product(6, "Wrist watch", "Accessories", 10, 99.99)
        ]
        # Insert each default product into the BST.
        for product in products:
            self.inventory.insert(product)

    def display_products(self):
        """Display all products in sorted order."""
        products = self.inventory.inorder()  # Retrieve products sorted by ID.
        # Print a header for the product list.
        print(f"\n{'ID':<5} {'Name':<15} {'Category':<15} {'Quantity':<10} {'Price':<10}")
        print("-" * 55)  # Print a separator line.
        for product in products:
            # Print each product's details in a formatted table row.
            print(f"{product.id:<5} {product.name:<15} {product.category:<15} {product.quantity:<10} ${product.price:<10.2f}")

    def add_product(self):
        """Add a new product to the inventory."""
        try:
            id = int(input("Enter Product ID: "))  # Prompt user for product ID.
            # Check if a product with this ID already exists.
            if self.inventory.search_by_id(id):
                print("A product with this ID already exists!")
                return
            name = input("Enter Product Name: ")         # Get product name.
            category = input("Enter Product Category: ")   # Get product category.
            quantity = int(input("Enter Product Quantity: "))  # Get available quantity.
            price = float(input("Enter Product Price: "))      # Get product price.
            new_product = Product(id, name, category, quantity, price)  # Create a new Product.
            self.inventory.insert(new_product)  # Insert the new product into the BST.
            print("Product added successfully!")
        except ValueError:
            # Handle invalid inputs (e.g., non-numeric values).
            print("Invalid input! Please try again.")

    def delete_product(self):
        """Delete an existing product by ID."""
        try:
            id = int(input("Enter Product ID to delete: "))  # Prompt user for the product ID.
            node = self.inventory.search_by_id(id)  # Search for the product in the BST.
            if node:
                self.inventory.delete(id)  # Delete the product if it exists.
                print("Product deleted successfully!")
            else:
                print("Product not found!")
        except ValueError:
            print("Invalid input! Please enter a valid number.")

    def buy_product(self):
        """Add a product to the shopping cart after checking stock availability."""
        identifier = input("Enter Product Name or ID: ")  # Get product identifier (name or id).
        node = None
        if identifier.isdigit():
            # If the identifier is numeric, search by ID.
            node = self.inventory.search_by_id(int(identifier))
        else:
            # Otherwise, search by name.
            products = self.inventory.search_by_name(identifier)
            if products:
                node = self.inventory.search_by_id(products[0].id)
        
        if not node:
            print("Product not found!")
            return
        
        try:
            quantity = int(input("Enter quantity: "))  # Ask user for the desired quantity.
            if quantity > node.product.quantity:
                # Check if the requested quantity is available.
                print("Not enough stock!")
                return
            # Add the product and the requested quantity to the shopping cart.
            self.cart.append((node.product, quantity))
            node.product.quantity -= quantity  # Update the product's stock quantity.
            print("Product added to cart!")
        except ValueError:
            print("Invalid quantity!")

    def checkout(self):
        """Calculate the total cost, apply tax, and clear the shopping cart."""
        if not self.cart:
            print("Cart is empty!")
            return
        # Calculate the total cost of items in the cart.
        total = sum(product.price * quantity for product, quantity in self.cart)
        tax = total * 0.07  # Calculate tax at 7%.
        final_total = total + tax  # Compute the final total.
        print("\n===== Checkout =====")
        print(f"Total Price: ${total:.2f}")    # Display total price.
        print(f"Tax (7%): ${tax:.2f}")          # Display calculated tax.
        print(f"Final Total: ${final_total:.2f}") # Display final amount to pay.
        self.cart.clear()  # Clear the shopping cart after checkout.

    def menu(self):
        """Display the interactive menu for the inventory system."""
        while True:
            # Print the menu options for the user.
            print("\n===== Inventory Menu =====")
            print("1. View All Products")
            print("2. Search Product by ID")
            print("3. Search Product by Category")
            print("4. Add Product")
            print("5. Delete Product")
            print("6. Buy Product")
            print("7. Checkout")
            print("8. Run Test Cases")  # New option to run test cases.
            print("9. Exit")
            try:
                choice = int(input("Enter your choice: "))  # Get the user's menu choice.
                if choice == 1:
                    self.display_products()  # Option 1: Display all products.
                elif choice == 2:
                    try:
                        id = int(input("Enter Product ID: "))  # Option 2: Search by product ID.
                        node = self.inventory.search_by_id(id)
                        if node:
                            print(vars(node.product))  # Display the product details.
                        else:
                            print("Product not found!")
                    except ValueError:
                        print("Invalid input! Please enter a number.")
                elif choice == 3:
                    category = input("Enter Product Category: ")  # Option 3: Search by category.
                    products = self.inventory.search_by_category(category)
                    if products:
                        for product in products:
                            print(vars(product))  # Display matching products.
                    else:
                        print("No products found in this category!")
                elif choice == 4:
                    self.add_product()  # Option 4: Add a new product.
                elif choice == 5:
                    self.delete_product()  # Option 5: Delete a product.
                elif choice == 6:
                    self.buy_product()  # Option 6: Buy a product.
                elif choice == 7:
                    self.checkout()  # Option 7: Checkout the cart.
                elif choice == 8:
                    run_tests()  # Option 8: Run the test cases.
                elif choice == 9:
                    print("Exiting...")  # Option 9: Exit the program.
                    break
                else:
                    print("Invalid choice! Please try again.")
            except ValueError:
                # Handle non-numeric input for the menu choice.
                print("Invalid input! Please enter a number.")

# -------------------------------
# Main execution: interactive menu or test suite
# -------------------------------

def main():
    # Create an instance of the InventorySystem.
    system = InventorySystem()
    # Launch the interactive menu for user operations.
    system.menu()

if __name__ == "__main__":
    # If the script is called with "test" as a command-line argument, run tests.
    if len(sys.argv) > 1 and sys.argv[1] == "test":
        unittest.main(argv=[sys.argv[0]])
    else:
        main()
```
